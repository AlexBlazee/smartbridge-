```python
Z = linkage(df,'ward')
plt.figure(figsize=(50,10))
plt.title('Hierarchial Clustering Dendrogram')
plt.xlabel('Iris Dataset')
plt.ylabel('distance')
dendrogram(
    Z,
    labels = labels,
leaf_rotation = 90.,
leaf_font_size =8.,
)
plt.show()
```
